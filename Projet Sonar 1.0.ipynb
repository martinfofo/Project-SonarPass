```python
def Passer(player):

    local_df = df.copy(deep=True)

    local_df = local_df[local_df["StatName"]=="Passe"]

    local_df = local_df[local_df["PlayerName"]==player]

    local_df = local_df.dropna(axis=1, how="all")



    df1 = local_df[['AnglePasse','LongueurPasse']].copy()



    bins = np.linspace(-np.pi,np.pi,10)



    df1['binned'] = pd.cut(local_df['AnglePasse'], bins, include_lowest=True, right = True)

    df1["Bin_Mids"] = df1["binned"].apply(lambda x: x.mid)

    df1 = df1[:-1]



    A= df1.groupby("Bin_Mids", as_index=False)["LongueurPasse"].mean()

    A = A.dropna(0)



    return A
    print(bins)
```
